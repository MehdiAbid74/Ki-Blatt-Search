Spielehalle

```python
class Objekt:
    def __init__(self, name, mas, position):
        self.name = name          # Name des Objekts
        self.mas = mas            # [höhe, breite]
        self.position = position  # [zeile, spalte]

# Spielfeldgröße: 10 x 6
field_size = [10, 6]

# Liste mit Objekten
objekte = [
    Objekt("A Bar", [3, 4], [0, 0]),
    Objekt("B Berg", [5, 4], [0, 0])
]

def print_field(objekte):
    print("--- --- ---")
    rows, cols = field_size
    grid = [["." for _ in range(cols)] for _ in range(rows)]

    # Jedes Objekt ins Raster einzeichnen (mit dem ersten Buchstaben)
    for o in objekte:
        r0, c0 = o.position
        h, w = o.mas
        for r in range(r0, min(rows, r0 + h)):
            for c in range(c0, min(cols, c0 + w)):
                grid[r][c] = o.name[0].upper()

    # Ausgabe
    for row in grid:
        print(" ".join(row))

print_field(objekte)
```

```python
# Constrains

#abstand min 1
def c1(x: Objekt, y: Objekt, s: int = 1) -> bool:
    r1, c1 = x.position
    h1, w1 = x.mas
    r2, c2 = y.position
    h2, w2 = y.mas

    # Trennung in Zeilen (x oberhalb von y oder x unterhalb von y)
    row_separated = (r1 + h1 + s <= r2) or (r2 + h2 + s <= r1)

    # Trennung in Spalten (x links von y oder x rechts von y)
    col_separated = (c1 + w1 + s <= c2) or (c2 + w2 + s <= c1)

    # Wenn entweder in Zeilen ODER in Spalten getrennt, dann kein Konflikt
    return row_separated or col_separated

c1(objekte[0],objekte[1])
```

```python
#abtand genau 1, berg bar , bar eingang
def c3(berg: Objekt,bar: Objekt):
    return c2(berg,bar)

def c4(bar: Objekt,eingang: Objekt):
    return c2(bar,eingang)

def c2(x: Objekt, y: Objekt) -> bool:
    r1, c1 = x.position
    h1, w1 = x.mas
    r2, c2 = y.position
    h2, w2 = y.mas

    # horizontal nebeneinander mit Abstand genau 1
    horizontal_adj = (c1 + w1 + 1 == c2 or c2 + w2 + 1 == c1) and not (r1 + h1 - 1 < r2 or r2 + h2 - 1 < r1)

    # vertikal nebeneinander mit Abstand genau 1
    vertical_adj = (r1 + h1 + 1 == r2 or r2 + h2 + 1 == r1) and not (c1 + w1 - 1 < c2 or c2 + w2 - 1 < c1)

    return horizontal_adj or vertical_adj
```
