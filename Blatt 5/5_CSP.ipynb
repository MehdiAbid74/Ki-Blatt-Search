```python
class Objekt:
    def __init__(self,  mas,pos):
        self.mas = mas
        self.pos = pos

mase = []
xrange = 10
yrange = 6
for x in range(xrange):
    for y in range(yrange):
        mase.append((x, y))

berg = Objekt([4,5],mase)
bar = Objekt([3,2],mase)
eing = Objekt([1,1],mase)
naus = Objekt([1,1],mase)

def printAllMase():
    print(f"Berg: {berg.pos}")
    print(f"Bar: {bar.pos}")
    print(f"Eing: {eing.pos}")
    print(f"Naus: {naus.pos}")

printAllMase()
```

start bei bar, da am meisten constrains

c1: (x element Objekte, y element Objekte) x y abstand min 1


c2: abstand genau 1 (bar,eing) (bar,berg)

c3 türen am äußerem rand (wand)

"Ohne drehen zunächst"

wenn wir den berg bei {0,0} hinstellen würde er von v2(berg) nichts ändern

das würd sich so lange durchziehen bis wir bei den türen sind die als constrain haben an der wand zu stehen:

```python
def cwand(o: Objekt):
    newpos = []
    for p in o.pos:
        x, y = p
        if x == 0 or x == xrange - 1 or y == 0 or y == yrange - 1:
            newpos.append(p)
    o.pos = newpos

cwand(eing)
```

```python

printAllMase()
```

```python
cwand(naus)
printAllMase()
```
