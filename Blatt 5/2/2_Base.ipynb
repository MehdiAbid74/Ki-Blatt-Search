Implementation of CSP, for riddling the einstein problem

Es gibt fünf Häuser.


Der Engländer wohnt im roten Haus.


Der Spanier hat einen Hund.


Kaffee wird im grünen Haus getrunken.


Der Ukrainer trinkt Tee.


Das grüne Haus ist direkt rechts vom weißen Haus.


Der Raucher von Old-Gold-Zigaretten hält Schnecken als Haustiere.


Die Zigaretten der Marke Kools werden im gelben Haus geraucht.


Milch wird im mittleren Haus getrunken.


Der Norweger wohnt im ersten Haus.


Der Mann, der Chesterfield raucht, wohnt neben dem Mann mit dem Fuchs.


Die Marke Kools wird geraucht im Haus neben dem Haus mit dem Pferd.


Der Lucky-Strike-Raucher trinkt am liebsten Orangensaft.


Der Japaner raucht Zigaretten der Marke Parliaments.


Der Norweger wohnt neben dem blauen Haus.


Wer trinkt Wasser? Wem gehört das Zebra?

```python
from enum import Enum

class Farbe(Enum):
    ROT = "rot"
    GRUEN = "grün"
    WEISS = "weiß"
    BLAU = "blau"
    GELB = "gelb"
alleFarben = [Farbe.ROT, Farbe.GRUEN,Farbe.WEISS,Farbe.BLAU,Farbe.GELB]

class Nationalitaet(Enum):
    ENGLAENDER = "Engländer"
    UKRAINER = "Ukrainer"
    SPANIER = "Spanier"
    NORWEGER = "Norweger"
    JAPANER = "Japaner"
alleNationalitaten = [Nationalitaet.ENGLAENDER, Nationalitaet.UKRAINER,Nationalitaet.SPANIER,Nationalitaet.NORWEGER,Nationalitaet.JAPANER]

class Getraenk(Enum):
    TEE = "Tee"
    KAFFEE = "Kaffee"
    MILCH = "Milch"
    ORANGENSAFT = "Orangensaft"
    WASSER = "Wasser"
alleGetranke = [Getraenk.TEE,Getraenk.KAFFEE,Getraenk.MILCH,Getraenk.ORANGENSAFT,Getraenk.WASSER]

class Zigarette(Enum):
    KOOLS = "Kools"
    LUCKY = "Lucky"
    PARLIAMENTS = "Parliaments"
    CHESFIELD = "Chesfield"
    OLD_GOLD = "Old_Gold"
alleZigaretten = [Zigarette.KOOLS,Zigarette.LUCKY,Zigarette.PARLIAMENTS,Zigarette.CHESFIELD,Zigarette.OLD_GOLD]

class Tier(Enum):
    ZEBRA = "Zebra"
    HUND = "Hund"
    SCHNECKE = "Schnecke"
    PFERD = "Pferd"
    FUCHS = "Fuchs"
alleTiere = [Tier.ZEBRA,Tier.HUND,Tier.SCHNECKE,Tier.PFERD,Tier.FUCHS]
```

```python
import pandas as pd

# Kategorien und Werte



# Anzahl der Häuser
n = 5

class Haus:
    def __init__(self, farbe=None, nationalitaet=None, getraenk=None, zigarette=None, tier=None):
        self.farbe = farbe
        self.nationalitaet = nationalitaet
        self.getraenk = getraenk
        self.zigarette = zigarette
        self.tier = tier

    def __repr__(self):
        return f"Haus(farbe={self.farbe}, nat={self.nationalitaet}, getr={self.getraenk}, zig={self.zigarette}, tier={self.tier})"

#häuser erstellen
haeuser = [Haus() for _ in range(5)] #haus 1 = index 1

# Name und Position hinzufügen
for idx, haus in enumerate(haeuser, 1):
    haus.name = f"Haus {idx}"
    haus.position = idx

def alleWerteHinzufügen():
    for haus in haeuser:
        haus.farbe = alleFarben
        haus.tier = alleTiere
        haus.getraenk = alleGetranke
        haus.zigarette = alleZigaretten
        haus.nationalitaet = alleNationalitaten

def c9(haus:Haus):
    if haus.name == 3:
        haus.getraenk = Getraenk.MILCH
def c10(haus:Haus):
    if haus.name == 1:
        haus.nationalitaet = Nationalitaet.NORWEGER
#constrains:
def c1(haus):
    # Engländer ⇔ rot
    iff(haus.nationalitaet, Nationalitaet.ENGLAENDER,
        haus.farbe, Farbe.ROT)
def c2(haus):
    # Spanier ⇔ Hund
    iff(haus.nationalitaet, Nationalitaet.SPANIER,
        haus.tier, Tier.HUND)
def c3(haus):
    # Kaffee ⇔ Grün
    iff(haus.getraenk, Getraenk.KAFFEE,
        haus.farbe, Farbe.GRUEN)
def c4(haus):
    # Ukrainer ⇔ Tee
    iff(haus.nationalitaet, Nationalitaet.UKRAINER,
        haus.getraenk, Getraenk.TEE)
def c5(haus):
    # Old-Gold ⇔ Schnecke
    iff(haus.zigarette, Zigarette.OLD_GOLD,
        haus.tier, Tier.SCHNECKE)
def c6(haus):
    # Kools ⇔ Gelb
    iff(haus.zigarette, Zigarette.KOOLS,
        haus.farbe, Farbe.GELB)
def c7(haus):
    # Lucky ⇔ Orangensaft
    iff(haus.zigarette, Zigarette.LUCKY,
        haus.getraenk, Getraenk.ORANGENSAFT)
def c8(haus):
    # Japaner ⇔ Parliaments
    iff(haus.nationalitaet, Nationalitaet.JAPANER,
        haus.zigarette, Zigarette.PARLIAMENTS)

def iff(a_set, a_value, b_set, b_value):
    # a ⇒ b
    if a_value in a_set and b_value not in b_set:
        a_set.remove(a_value)

    # b ⇒ a
    if b_value in b_set and a_value not in a_set:
        b_set.remove(b_value)

# binäre
def sindNachbarn(haus1, haus2):
    return abs(haus1.position - haus2.position) == 1


def remove_if_no_match(set1, val1, set2, val2):
    """Entfernt val1 aus set1, wenn val2 nicht in set2 ist."""
    if val1 in set1 and val2 not in set2:
        set1.remove(val1)
        return True
    return False

def c11(haus1, haus2):
    """Chesterfield-Nachbar mit Fuchs"""
    if not sindNachbarn(haus1, haus2):
        return False  # Häuser nicht nebeneinander → nichts tun

    changed = False
    # Wenn haus1 Chesfield raucht, haus2 muss Fuchs haben (Nachbar)
    changed |= remove_if_no_match(haus1.zigarette, Zigarette.CHESFIELD, haus2.tier, Tier.FUCHS)
    changed |= remove_if_no_match(haus2.zigarette, Zigarette.CHESFIELD, haus1.tier, Tier.FUCHS)
    # Umgekehrt: wenn Haus1 Fuchs hat, Haus2 muss Chesfield haben (Nachbar)
    changed |= remove_if_no_match(haus1.tier, Tier.FUCHS, haus2.zigarette, Zigarette.CHESFIELD)
    changed |= remove_if_no_match(haus2.tier, Tier.FUCHS, haus1.zigarette, Zigarette.CHESFIELD)
    return changed

def c12(haus1, haus2):
    if not sindNachbarn(haus1, haus2):
        return False
    changed = False
    changed |= remove_if_no_match(haus1.zigarette, Zigarette.KOOLS, haus2.tier, Tier.PFERD)
    changed |= remove_if_no_match(haus2.zigarette, Zigarette.KOOLS, haus1.tier, Tier.PFERD)
    changed |= remove_if_no_match(haus1.tier, Tier.PFERD, haus2.zigarette, Zigarette.KOOLS)
    changed |= remove_if_no_match(haus2.tier, Tier.PFERD, haus1.zigarette, Zigarette.KOOLS)
    return changed

def c14(haus1, haus2):
    if not sindNachbarn(haus1, haus2):
        return False
    changed = False
    changed |= remove_if_no_match(haus1.nationalitaet, Nationalitaet.NORWEGER, haus2.farbe, Farbe.BLAU)
    changed |= remove_if_no_match(haus2.nationalitaet, Nationalitaet.NORWEGER, haus1.farbe, Farbe.BLAU)
    changed |= remove_if_no_match(haus1.farbe, Farbe.BLAU, haus2.nationalitaet, Nationalitaet.NORWEGER)
    changed |= remove_if_no_match(haus2.farbe, Farbe.BLAU, haus1.nationalitaet, Nationalitaet.NORWEGER)
    return changed

def c13(haus1, haus2):
    """Grünes Haus ist rechts vom weißen Haus"""
    if not sindNachbarn(haus1, haus2):
        return False
    changed = False
    # Haus1 links, Haus2 rechts
    if haus1.position < haus2.position:
        changed |= remove_if_no_match(haus1.farbe, Farbe.WEISS, haus2.farbe, Farbe.GRUEN)
        changed |= remove_if_no_match(haus2.farbe, Farbe.GRUEN, haus1.farbe, Farbe.WEISS)
    # Haus2 links, Haus1 rechts
    elif haus2.position < haus1.position:
        changed |= remove_if_no_match(haus2.farbe, Farbe.WEISS, haus1.farbe, Farbe.GRUEN)
        changed |= remove_if_no_match(haus1.farbe, Farbe.GRUEN, haus2.farbe, Farbe.WEISS)
    return changed

def printEnum(EnumClass, haus):
    attr_name = EnumClass.__name__.lower()
    value = getattr(haus, attr_name)

    # Fall 1: Einzel-Enum
    if isinstance(value, EnumClass):
        return value.name

    # Fall 2: Iterable (Liste)
    return " ".join(e.name for e in EnumClass if e in value)


# Test-Ausgabe
def printHäuser():
    for i, haus in enumerate(haeuser, 1):
        print(f"Haus {i}:")
        print(printEnum(Nationalitaet,haus))
        print(printEnum(Zigarette,haus))
        print(printEnum(Tier,haus))
        print(printEnum(Farbe,haus))
        print(printEnum(Getraenk,haus))


```

Unäre Constrains als erstes:

```python
def useConstrainOnAll(haeuser, constrain):
    for i, haus in enumerate(haeuser, 1):
        result = constrain(haus)

# die zwei gesetzen, köönnten osnt auch als unären aber so einfacher

def useAllBinärConstrains(haus1,haus2):
    change = False
    change |= c11(haus1, haus2)
    change |= c12(haus1, haus2)
    change |= c13(haus1, haus2)
    change |= c14(haus1, haus2)
    return change


def useAllUnärConstrains(haeuser):

    useConstrainOnAll(haeuser,c9)
    useConstrainOnAll(haeuser,c10)

    useConstrainOnAll(haeuser,c1)
    useConstrainOnAll(haeuser,c2)
    useConstrainOnAll(haeuser,c3)
    useConstrainOnAll(haeuser,c4)
    useConstrainOnAll(haeuser,c5)
    useConstrainOnAll(haeuser,c6)
    useConstrainOnAll(haeuser,c7)
    useConstrainOnAll(haeuser,c8)

```

```python
# Queue für binäre Constraints erstellen
queueMax = []
#for i in range(len(haeuser)):
#    for j in range(len(haeuser)):
#        if i != j:  # kein Haus mit sich selbst
#            queueMax.append((haeuser[i], haeuser[j]))
queueMax = []
for i in range(len(haeuser)-1):
    queueMax.append((haeuser[i], haeuser[i+1]))
    queueMax.append((haeuser[i+1], haeuser[i]))
queue = queueMax.copy()


# Kopie für die aktuelle Queue
queue = queueMax.copy()

def print_queue():
    print("Aktuelle Queue:")
    for q in queue:
        print(f"{q[0].name} -> {q[1].name}")

def pop_queue():
    if queue:
        popped = queue.pop(0)
        #print(f"Pop: {popped[0].name} -> {popped[1].name}")
        return popped
    else:
        print("Queue ist leer")
        return None

def reset_queue():
    global queue
    queue = queueMax.copy()

def queueAppendAll(o:Haus):
    for f in queueMax:
        if f[1] == o:
            queue.append(f)

print_queue()
```

```python
def arc3():
    # Unäre Constraints einmal zu Beginn anwenden
    reset_queue()
    useAllUnärConstrains(haeuser)

    # AC-3 Loop
    while queue:
        h1, h2 = pop_queue()
        if h1 is None or h2 is None:
            break  # Queue leer

        # Binäre Constraints auf h1, h2 anwenden
        changed = useAllBinärConstrains(h1, h2)

        # Wenn Domäne geändert, alle Nachbarn von h1 wieder in Queue
        if changed:
            queueAppendAll(h2)
            # Optional: unäre Constraints erneut auf alle Häuser anwenden
            useAllUnärConstrains(haeuser)

    print("AC-3 abgeschlossen")
arc3()
```
