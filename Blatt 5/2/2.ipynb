Implementation of CSP, for riddling the einstein problem

Es gibt fünf Häuser.


Der Engländer wohnt im roten Haus.


Der Spanier hat einen Hund.


Kaffee wird im grünen Haus getrunken.


Der Ukrainer trinkt Tee.


Das grüne Haus ist direkt rechts vom weißen Haus.


Der Raucher von Old-Gold-Zigaretten hält Schnecken als Haustiere.


Die Zigaretten der Marke Kools werden im gelben Haus geraucht.


Milch wird im mittleren Haus getrunken.


Der Norweger wohnt im ersten Haus.


Der Mann, der Chesterfield raucht, wohnt neben dem Mann mit dem Fuchs.


Die Marke Kools wird geraucht im Haus neben dem Haus mit dem Pferd.


Der Lucky-Strike-Raucher trinkt am liebsten Orangensaft.


Der Japaner raucht Zigaretten der Marke Parliaments.


Der Norweger wohnt neben dem blauen Haus.


Wer trinkt Wasser? Wem gehört das Zebra?

```python
from enum import Enum
class Farbe(Enum):
    ROT = "rot"
    GRUEN = "grün"
    WEISS = "weiß"
    BLAU = "blau"
    GELB = "gelb"

class Nationalitaet(Enum):
    ENGLAENDER = "Engländer"
    UKRAINER = "Ukrainer"
    SPANIER = "Spanier"
    NORWEGER = "Norweger"
    JAPANER = "Japaner"

class Getraenk(Enum):
    TEE = "Tee"
    KAFFEE = "Kaffee"
    MILCH = "Milch"
    ORANGENSAFT = "Orangensaft"
    WASSER = "Wasser"

class Zigarette(Enum):
    KOOLS = "Kools"
    LUCKY = "Lucky"
    PARLIAMENTS = "Parliaments"
    CHESFIELD = "Chesfield"
    OLD_GOLD = "Old_Gold"

class Tier(Enum):
    ZEBRA = "Zebra"
    HUND = "Hund"
    SCHNECKE = "Schnecke"
    PFERDE = "Pferde"
    FUCHS = "Fuchs"
```

```python
import pandas as pd

# Kategorien und Werte



# Anzahl der Häuser
n = 5

class Haus:
    def __init__(self, farbe=None, nationalitaet=None, getraenk=None, zigarette=None, tier=None):
        self.farbe = farbe
        self.nationalitaet = nationalitaet
        self.getraenk = getraenk
        self.zigarette = zigarette
        self.tier = tier

    def __repr__(self):
        return f"Haus(farbe={self.farbe}, nat={self.nationalitaet}, getr={self.getraenk}, zig={self.zigarette}, tier={self.tier})"

#häuser erstellen
haeuser = [Haus() for _ in range(5)]

#constrains:

# Unäre und binäre Constraints innerhalb eines Hauses
def c1(haus):
    # Der Engländer wohnt im roten Haus
    if haus.farbe is None or haus.nationalitaet is None:
        return True
    return haus.farbe == Farbe.ROT and haus.nationalitaet == Nationalitaet.ENGLAENDER

def c2(haus):
    # Der Spanier hat einen Hund
    if haus.nationalitaet is None or haus.tier is None:
        return True
    return haus.nationalitaet == Nationalitaet.SPANIER and haus.tier == Tier.HUND

def c3(haus):
    # Kaffee wird im grünen Haus getrunken
    if haus.getraenk is None or haus.farbe is None:
        return True
    return haus.getraenk == Getraenk.KAFFEE and haus.farbe == Farbe.GRUEN

def c4(haus):
    # Der Ukrainer trinkt Tee
    if haus.nationalitaet is None or haus.getraenk is None:
        return True
    return haus.nationalitaet == Nationalitaet.UKRAINER and haus.getraenk == Getraenk.TEE

def c5(haus):
    # Old-Gold-Raucher hat Schnecken
    if haus.zigarette is None or haus.tier is None:
        return True
    return haus.zigarette == Zigarette.OLD_GOLD and haus.tier == Tier.SCHNECKE

def c6(haus):
    # Kools im gelben Haus
    if haus.zigarette is None or haus.farbe is None:
        return True
    return haus.zigarette == Zigarette.KOOLS and haus.farbe == Farbe.GELB

def c7(haus):
    # Lucky trinkt Orangensaft
    if haus.zigarette is None or haus.getraenk is None:
        return True
    return haus.zigarette == Zigarette.LUCKY and haus.getraenk == Getraenk.ORANGENSAFT

def c8(haus):
    # Japaner raucht Parliaments
    if haus.nationalitaet is None or haus.zigarette is None:
        return True
    return haus.nationalitaet == Nationalitaet.JAPANER and haus.zigarette == Zigarette.PARLIAMENTS

# Unäre Constraints für bestimmte Positionen
def c9(haeuser):
    # Milch im mittleren Haus (Haus 3, Index 2)
    return haeuser[2].getraenk == Getraenk.MILCH

def c10(haeuser):
    # Norweger im ersten Haus (Haus 1, Index 0)
    return haeuser[0].nationalitaet == Nationalitaet.NORWEGER

# Nachbarschafts-Constraints zwischen Häusern
def c11(haeuser):
    # Chesterfield-Nachbar mit Fuchs
    for i, haus in enumerate(haeuser):
        if haus.zigarette == Zigarette.CHESFIELD:
            left = i > 0 and haeuser[i-1].tier == Tier.FUCHS
            right = i < len(haeuser)-1 and haeuser[i+1].tier == Tier.FUCHS
            return left or right
    return True

def c12(haeuser):
    # Kools-Nachbar mit Pferd
    for i, haus in enumerate(haeuser):
        if haus.zigarette == Zigarette.KOOLS:
            left = i > 0 and haeuser[i-1].tier == Tier.PFERDE
            right = i < len(haeuser)-1 and haeuser[i+1].tier == Tier.PFERDE
            return left or right
    return True

def c13(haeuser):
    # Grünes Haus ist rechts vom weißen Haus
    pos_gruen = pos_weiss = None
    for i, haus in enumerate(haeuser):
        if haus.farbe == Farbe.GRUEN:
            pos_gruen = i
        if haus.farbe == Farbe.WEISS:
            pos_weiss = i
    if pos_gruen is None or pos_weiss is None:
        return True
    return pos_gruen == pos_weiss + 1

def c14(haeuser):
    # Norweger neben blauem Haus
    for i, haus in enumerate(haeuser):
        if haus.nationalitaet == Nationalitaet.NORWEGER:
            left = i > 0 and haeuser[i-1].farbe == Farbe.BLAU
            right = i < len(haeuser)-1 and haeuser[i+1].farbe == Farbe.BLAU
            return left or right
    return True





# Test-Ausgabe
def printHäuser():
    for i, haus in enumerate(haeuser, 1):
        print(f"Haus {i}: {haus}")

```
