Implementation of CSP, for riddling the einstein problem

Es gibt fünf Häuser.


Der Engländer wohnt im roten Haus.


Der Spanier hat einen Hund.


Kaffee wird im grünen Haus getrunken.


Der Ukrainer trinkt Tee.


Das grüne Haus ist direkt rechts vom weißen Haus.


Der Raucher von Old-Gold-Zigaretten hält Schnecken als Haustiere.


Die Zigaretten der Marke Kools werden im gelben Haus geraucht.


Milch wird im mittleren Haus getrunken.


Der Norweger wohnt im ersten Haus.


Der Mann, der Chesterfield raucht, wohnt neben dem Mann mit dem Fuchs.


Die Marke Kools wird geraucht im Haus neben dem Haus mit dem Pferd.


Der Lucky-Strike-Raucher trinkt am liebsten Orangensaft.


Der Japaner raucht Zigaretten der Marke Parliaments.


Der Norweger wohnt neben dem blauen Haus.


Wer trinkt Wasser? Wem gehört das Zebra?

```python
from enum import Enum

class Farbe(Enum):
    ROT = "rot"
    GRUEN = "grün"
    WEISS = "weiß"
    BLAU = "blau"
    GELB = "gelb"
alleFarben = [Farbe.ROT, Farbe.GRUEN,Farbe.WEISS,Farbe.BLAU,Farbe.GELB]

class Nationalitaet(Enum):
    ENGLAENDER = "Engländer"
    UKRAINER = "Ukrainer"
    SPANIER = "Spanier"
    NORWEGER = "Norweger"
    JAPANER = "Japaner"
alleNationalitaten = [Nationalitaet.ENGLAENDER, Nationalitaet.UKRAINER,Nationalitaet.SPANIER,Nationalitaet.NORWEGER,Nationalitaet.JAPANER]

class Getraenk(Enum):
    TEE = "Tee"
    KAFFEE = "Kaffee"
    MILCH = "Milch"
    ORANGENSAFT = "Orangensaft"
    WASSER = "Wasser"
alleGetranke = [Getraenk.TEE,Getraenk.KAFFEE,Getraenk.MILCH,Getraenk.ORANGENSAFT,Getraenk.WASSER]

class Zigarette(Enum):
    KOOLS = "Kools"
    LUCKY = "Lucky"
    PARLIAMENTS = "Parliaments"
    CHESFIELD = "Chesfield"
    OLD_GOLD = "Old_Gold"
alleZigaretten = [Zigarette.KOOLS,Zigarette.LUCKY,Zigarette.PARLIAMENTS,Zigarette.CHESFIELD,Zigarette.OLD_GOLD]

class Tier(Enum):
    ZEBRA = "Zebra"
    HUND = "Hund"
    SCHNECKE = "Schnecke"
    PFERDE = "Pferde"
    FUCHS = "Fuchs"
alleTiere = [Tier.ZEBRA,Tier.HUND,Tier.SCHNECKE,Tier.PFERDE,Tier.FUCHS]
```

```python
import pandas as pd

# Kategorien und Werte



# Anzahl der Häuser
n = 5

class Haus:
    def __init__(self, farbe=None, nationalitaet=None, getraenk=None, zigarette=None, tier=None):
        self.farbe = farbe
        self.nationalitaet = nationalitaet
        self.getraenk = getraenk
        self.zigarette = zigarette
        self.tier = tier

    def __repr__(self):
        return f"Haus(farbe={self.farbe}, nat={self.nationalitaet}, getr={self.getraenk}, zig={self.zigarette}, tier={self.tier})"

#häuser erstellen
haeuser = [Haus() for _ in range(5)] #haus 1 = index 1


def alleWerteHinzufügen():
    for haus in haeuser:
        haus.farbe = alleFarben
        haus.tier = alleTiere
        haus.getraenk = alleGetranke
        haus.zigarette = alleZigaretten
        haus.nationalitaet = alleNationalitaten

#constrains:

# Unäre Constraints für bestimmte Positionen
def c9(haeuser):
    # Milch im mittleren Haus (Haus 3, Index 2)
    for i, haus in enumerate(haeuser, 1):
        if i == 3:
            haus.getraenk = Getraenk.MILCH

def c10(haeuser):
    # Norweger im ersten Haus (Haus 1, Index 0)
    for i, haus in enumerate(haeuser,1):
        if i == 1:
            haus.nationalitaet = Nationalitaet.NORWEGER


def c1(haus):
    # Der Engländer wohnt im roten Haus
    if haus.farbe == Farbe.ROT:
        haus.nationalitaet = Nationalitaet.ENGLAENDER
    elif haus.nationalitaet == Nationalitaet.ENGLAENDER:
        haus.farbe = Farbe.ROT


def c2(haus):
    # Der Spanier hat einen Hund
    if haus.nationalitaet == Nationalitaet.SPANIER:
        haus.tier = Tier.HUND
    elif haus.tier == Tier.HUND:
        haus.nationalitaet = Nationalitaet.SPANIER

def c3(haus):
    # Kaffee wird im grünen Haus getrunken
    if haus.getraenk == Getraenk.KAFFEE:
        haus.farbe = Farbe.GRUEN
    elif haus.farbe == Farbe.GRUEN:
        haus.getraenk = Getraenk.KAFFEE

def c4(haus):
    # Der Ukrainer trinkt Tee
    if haus.nationalitaet == Nationalitaet.UKRAINER:
        haus.getraenk = Getraenk.TEE
    elif haus.getraenk == Getraenk.TEE:
        haus.nationalitaet = Nationalitaet.UKRAINER

def c5(haus):
    # Old-Gold-Raucher hat Schnecken
    if haus.zigarette == Zigarette.OLD_GOLD:
        haus.tier = Tier.SCHNECKE
    elif haus.tier == Tier.SCHNECKE:
        haus.zigarette = Zigarette.OLD_GOLD

def c6(haus):
    # Kools im gelben Haus
    if haus.zigarette == Zigarette.KOOLS:
        haus.farbe = Farbe.GELB
    elif haus.farbe == Farbe.GELB:
        haus.zigarette = Zigarette.KOOLS

def c7(haus):
    # Lucky trinkt Orangensaft
    if haus.zigarette == Zigarette.LUCKY:
        haus.getraenk = Getraenk.ORANGENSAFT
    elif haus.getraenk == Getraenk.ORANGENSAFT:
        haus.zigarette = Zigarette.LUCKY

def c8(haus):
    # Japaner raucht Parliaments
    if haus.nationalitaet == Nationalitaet.JAPANER:
        haus.zigarette = Zigarette.PARLIAMENTS
    elif haus.zigarette == Zigarette.PARLIAMENTS:
        haus.nationalitaet = Nationalitaet.JAPANER

#binär

# Nachbarschafts-Constraints zwischen Häusern
def c11(haeuser):
    # Chesterfield-Nachbar mit Fuchs
    for i, haus in enumerate(haeuser):
        if haus.zigarette == Zigarette.CHESFIELD:
            left = i > 0 and haeuser[i-1].tier == Tier.FUCHS
            right = i < len(haeuser)-1 and haeuser[i+1].tier == Tier.FUCHS
            return left or right
    return True

def c12(haeuser):
    # Kools-Nachbar mit Pferd
    for i, haus in enumerate(haeuser):
        if haus.zigarette == Zigarette.KOOLS:
            left = i > 0 and haeuser[i-1].tier == Tier.PFERDE
            right = i < len(haeuser)-1 and haeuser[i+1].tier == Tier.PFERDE
            return left or right
    return True

def c13(haeuser):
    # Grünes Haus ist rechts vom weißen Haus
    pos_gruen = pos_weiss = None
    for i, haus in enumerate(haeuser):
        if haus.farbe == Farbe.GRUEN:
            pos_gruen = i
        if haus.farbe == Farbe.WEISS:
            pos_weiss = i
    if pos_gruen is None or pos_weiss is None:
        return True
    return pos_gruen == pos_weiss + 1

def c14(haeuser):
    # Norweger neben blauem Haus
    for i, haus in enumerate(haeuser):
        if haus.nationalitaet == Nationalitaet.NORWEGER:
            left = i > 0 and haeuser[i-1].farbe == Farbe.BLAU
            right = i < len(haeuser)-1 and haeuser[i+1].farbe == Farbe.BLAU
            return left or right
    return True

# Test-Ausgabe
def printHäuser():
    for i, haus in enumerate(haeuser, 1):
        print(f"Haus {i}: {haus}\n")

```

```python
def useConstrainOnAll(haeuser, constrain):
    for i, haus in enumerate(haeuser, 1):
        result = constrain(haus)

c9(haeuser)
c10(haeuser)

def useAllUnärConstrains(haeuser):
    useConstrainOnAll(haeuser,c1)
    useConstrainOnAll(haeuser,c2)
    useConstrainOnAll(haeuser,c3)
    useConstrainOnAll(haeuser,c4)
    useConstrainOnAll(haeuser,c5)
    useConstrainOnAll(haeuser,c6)
    useConstrainOnAll(haeuser,c7)
    useConstrainOnAll(haeuser,c8)

def useAllBinConstrains(haeuser):
    c11(haeuser)
    c12(haeuser)
    c13(haeuser)
    c14(haeuser)


useAllUnärConstrains(haeuser)
useAllBinConstrains(haeuser)

for i in range(5):
    useAllUnärConstrains(haeuser)
    useAllBinConstrains(haeuser)

printHäuser()
```
