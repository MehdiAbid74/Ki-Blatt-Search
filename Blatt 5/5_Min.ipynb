```python
class Objekt:
    def __init__(self,  mas,moglichePos,pos):
        self.mas = mas
        self.moglichePos = moglichePos
        self.pos = pos

mase = []
xrange = 10
yrange = 6
for x in range(xrange):
    for y in range(yrange):
        mase.append((x, y))

berg = Objekt([4,5],mase,(0,0))
bar = Objekt([3,2],mase,(0,0))
eing = Objekt([1,1],mase,(5,0))
naus = Objekt([1,1],mase,(0,4))

def printAllMase():
    print(f"Berg: {berg.pos} {berg.moglichePos}")
    print(f"Bar: {bar.pos} {bar.moglichePos}")
    print(f"Eing: {eing.pos} {eing.moglichePos}")
    print(f"Naus: {naus.pos} {naus.moglichePos}")

def cwand(o: Objekt):
    newpos = []
    for p in o.moglichePos:
        x, y = p
        if x == 0 or x == xrange - 1 or y == 0 or y == yrange - 1:
            newpos.append(p)
    o.moglichePos = newpos

cwand(eing)
cwand(naus)

# Funktion zur Konfliktprüfung (nur Beispiel für Sicherheitsabstand)
def conflict(o1, o2):
    r1, c1 = o1.pos
    h1, w1 = o1.mas
    r2, c2 = o2.pos
    h2, w2 = o2.mas
    s = 1  # Abstand
    row_separated = (r1 + h1 + s <= r2) or (r2 + h2 + s <= r1)
    col_separated = (c1 + w1 + s <= c2) or (c2 + w2 + s <= c1)
    return not (row_separated or col_separated)
# Start: zähle Konflikte pro Objekt
def total_conflicts(o, objekte):
    return sum(conflict(o, other) for other in objekte if other != o)

def valid_positions(o: Objekt, andere: Objekt, s=1):
    """Gibt alle Positionen in o.moglichePos zurück, die für andere direkt angrenzend (Abstand s) sind."""
    gültig = []
    h1, w1 = o.mas
    h2, w2 = andere.mas

    for r1, c1 in o.moglichePos:
        r2, c2 = andere.pos  # andere hat schon pos zugewiesen
        # horizontal nebeneinander mit Abstand s
        horizontal_adj = (c1 + w1 + s == c2 or c2 + w2 + s == c1) and not (r1 + h1 - 1 < r2 or r2 + h2 - 1 < r1)
        # vertikal nebeneinander mit Abstand s
        vertical_adj = (r1 + h1 + s == r2 or r2 + h2 + s == r1) and not (c1 + w1 - 1 < c2 or c2 + w2 - 1 < c1)

        if horizontal_adj or vertical_adj:
            gültig.append((r1, c1))

    return gültig

def non_adjacent_positions(o: Objekt, andere: Objekt, s=1):
    """Gibt alle Positionen in o.moglichePos zurück, die NICHT direkt angrenzend (Abstand > s) zu 'andere' sind."""
    gültig = []
    h1, w1 = o.mas
    h2, w2 = andere.mas

    for r1, c1 in o.moglichePos:
        r2, c2 = andere.pos  # andere hat schon pos zugewiesen

        # horizontal oder vertikal direkt angrenzend prüfen
        horizontal_adj = (c1 + w1 + s == c2 or c2 + w2 + s == c1) and not (r1 + h1 - 1 < r2 or r2 + h2 - 1 < r1)
        vertical_adj   = (r1 + h1 + s == r2 or r2 + h2 + s == r1) and not (c1 + w1 - 1 < c2 or c2 + w2 - 1 < c1)

        # nur Positionen, die NICHT angrenzend sind, behalten
        if not (horizontal_adj or vertical_adj):
            gültig.append((r1, c1))

    return gültig



printAllMase()
```
