(0.5P) Eine grundlegende Einschränkung des Perzeptrons besteht darin, dass es die EXKLUSIV-ODER-Funktion nicht implementieren kann. Erklären Sie den Grund für diese Einschränkung.
Thema: Verständnis Perzeptron

```python
import numpy as np
```

```python

def And(a,b):
    return a&b

def Or(a,b):
    return a|b

def Complement(a):
    return not a
```

```python
def activation(z):
    return 1 if z > 0 else -1
```

```python
def predict(x,w):
    z = np.dot(x,w)
    return activation(z)
```

```python
def train_perceptron(X,y,lr=0.5, max_stes=1000):
    wWeight = np.zeros(len(X[0]))

    for step in range(max_stes):
        errors = 0

        for xi,yi in zip(X,y):
            prediction_y = predict(xi,wWeight)
            if prediction_y != yi:
                #change weigts
                wWeight += lr * yi * xi
                #print(f"step: {step}, W: {wWeight}, yi: {yi}, prediction_y: {prediction_y}, xi: {xi}")
                errors += 1
        if errors == 0:
            break
    return wWeight
```

```python
# Beispiel-Daten für UND
X_bits = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
yAnd = np.array([-1, -1, -1, 1])
X = np.hstack([np.ones((X_bits.shape[0],1)), X_bits])  # Bias hinzufügen

yOr = np.array([-1, 1, 1, 1])

wAnd = train_perceptron(X, yAnd)
wOr = train_perceptron(X, yOr)


#bias = wo*1 verschiebung des ursprungs (0,0)
# Vorhersagen testen
def test(X,w):
    print(f"---Gewichte: {w}---")
    for xi in X:
        print(xi[1:], "->", predict(xi, w))
test(X,wAnd)
test(X,wOr)
```

```python
#Exklusiv or:

yXor = np.array([-1, 1, 1, -1])
wXor = train_perceptron(X, yXor)
test(X,wXor)
```

```python
Xnot = np.array([1,0]).reshape(-1,1)  # shape (2,1)
X = np.hstack([np.ones((Xnot.shape[0],1)), Xnot])  # Bias hinzufügen
ynot = np.array([-1, 1])  # KOMPLEMENT: 0->1, 1->-1

wNot = train_perceptron(X, ynot)
print(f"---Gewichte: {wNot}---")

for xi in X:
    print(xi[1:], "->", predict(xi, wNot))
```
